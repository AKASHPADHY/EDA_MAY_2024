- 1st convert the sseries into dictionalry

- when u convert it  becomes key:value pair

- iterate thorugh dictonary

- if that value== object type, then keep that in corresponding key in one list

- else keep that variable in other lsit

```python
data_types=visa_df.dtypes
x=dict(data_types)
x
```

```python
obj_li=[]
num_li=[]
for i,j in x.items():
    if j=='object':
        obj_li.append(i)
    else:
        num_li.append(i)
print(obj_li)
print(num_li)
```

```python
obj=[i  for i,j in x.items() if j=='object']
num=[i for i,j in x.items() if j!='object']
print(obj)
print(num)
```

$take$

- 1st argument: indices

- indices will take some values in array : in a list

- it tries to extract the data

- that values we provideed represents columns or rows??

- axis=0 represents about row

- axis=1 represents columns

- by default axis=0

```python
#case-1
visa_df.take([100,200,300])
#it retrives 3 rows which having index 100,200,300
```

```python
#case-2
visa_df.take([100,200,300],axis=1)
#now it will check for 100th ,200th,300th column
#but originally we have only 12 column(0,11)
#will get error
```

```python
#case-3 selecct 3 column 2nd,5th,7th
#using take
visa_df.take([2,5,7],axis=1)
```

```python
#case-4 slicing
#it will fail
visa_df.take([2:5],axis=1)
```
